```python
x0=np.ones((r,1))
print(np.shape(x0))
```

```python
x=np.transpose(x)
x0=np.transpose(x0)
x=np.concatenate((x,x0))
x=np.transpose(x)
print(np.shape(x))
```

```python
(r,c)=np.shape(x)
thetaall=[]

alpha=0.0001
j_cost=[]
for k in range(1,11):
    print(k)
    iter=0
    theta=np.zeros((c,1))
    temp=np.zeros((1,c))
    ytemp=np.zeros((r,1))
    ind=np.where(np.array(y)==k)
    ytemp[ind]=1
    while iter<1:
        dj=np.zeros((c,1))
        for j in range(0,c):
            for i in range(0,r):
                #print(np.shape(theta),np.shape([x[i]]))
                h=np.matmul(np.transpose(theta),np.transpose(x[i]))
                sigm_h=1/(1+math.exp(-h))
                dj[j]=dj[j] + ((ytemp[i]-sigm_h)*x[i][j])
            theta[j]=theta[j] + alpha*(dj[j])
        iter=iter+1
        thetaall.append(theta)
    
#print(np.shape(theta),np.shape(x[0]))
```

```python
#print(np.shape(thetaall))
thetaall=np.squeeze(thetaall)
print(np.shape(thetaall))

plt.plot(np.transpose(thetaall),"*")
plt.show()
print(thetaall)
```

```python
#print(np.shape(thetaall),np.shape([x[n,:]]))
n=3565
htest = np.matmul((x[n,:]),np.transpose(thetaall))
test_val=np.max(htest)
i =np.where(htest==test_val)
print(np.shape(i))
print(i)
print(np.shape(test_val))
print(test_val)
print("Actual and predicted y are: ",i[0]+1,y[n])
```
